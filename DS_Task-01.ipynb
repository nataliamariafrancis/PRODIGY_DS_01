```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#We skip the first four rows of the csv file since these rows contain metadata
worldbank = pd.read_csv(
    r"C:\Users\natal\Downloads\worldbank\API_SP.POP.TOTL_DS2_en_csv_v2_739890.csv",
    skiprows=4
)
worldbank.head()
```

```python
#We find the total population of each country that is listed in the year 2023
population_2023 = worldbank[['Country Name','2023']].copy()
population_2023.columns= 'Country Name','Total Population'
population_2023.dropna(inplace=True)
population_2023.head()
```

```python
plt.figure(figsize=(15,10))
sns.barplot(x='Total Population',y='Country Name',data=population_2023.sort_values(by="Total Population", ascending=True).head(25))
plt.title("Top 25 Countries with the least population in 2023")
plt.show()
```
